## Observe-t-on une corrélation entre le caratère bio des produits et la qualité nutritionnelle ?

```python
bio_quali = pd.merge(bio2, qualité_nutriti2, how='left', on='brands_tags')
bio_quali.head(20)
```

```python
X = "est_bio"
Y = "nutriscore_grade"

cont_1 = bio_quali[[X,Y]].dropna().pivot_table(index=X,columns=Y,aggfunc=len,margins=True,margins_name="Total")
cont_1
```

```python
tx = cont_1.loc[:,["Total"]]
ty = cont_1.loc[["Total"],:]
n = len(bio_quali[[X,Y]].dropna())
indep = round(tx.dot(ty) / n)
indep
```

```python
(cont_1 - indep)/indep
```

```python
from scipy.stats import chi2_contingency as chi2_contingency
```

```python
chi = chi2_contingency(cont_1)
chi, p, dof, expected_freq= chi2_contingency(cont_1)
```

```python
chi
```

```python
p
```

```python
import seaborn as sns


c = cont_1.fillna(0) # On remplace les valeurs nulles par 0
measure = (c-indep)**2/indep
xi_n = measure.sum().sum()
table = measure/xi_n
sns.heatmap(table.iloc[:-1,:-1],annot=c.iloc[:-1,:-1])
plt.title("Matrice de corrélation entre le bio et la qualité nutritive d'un produit")
plt.show()
```
